```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print("Question-1")
print("Description of the dataset is to describe the details of the house like id number, the year in which it is made,bedrooms along with washrooms and the price of the house according to specifications of the house")
data=pd.read_csv("G:\house.csv")
df=data.isnull().sum()
print(df)
```

```python
'''Here there is no missing value present in the dataset if there is any missing value present in the whole row or whole column then we will remove it as a whole by dropna() function or will find with imputation.In imputation we find out the mean,median of data which is numerical.We can find out imputation on any row we want or the same row and put the value of mean in the missing value place. We use mode as imputation for the categorical data and fill in the place by fillna() function'''
#Univariate graph
sns.distplot(data["price"], hist = True, kde = False)

```

```python
#Bivariate plot
sns.jointplot(x=data["price"], y=data['yr_built'], kind="scatter")
#price is independent on the year it is build. So no inference can be made
```

```python
sns.jointplot(data=data, x = "bedrooms", y = "price",kind = "scatter")

#Inference : price is not perfectly corelated with the number of bedrooms in the house, hence predictions cannot be made.
```

```python
sns.jointplot(data = data, x = "bathrooms", y = "price",kind = "scatter")

#INFERENCE : We get a randomised data with some outliers where price is not corelated with the number of bathrooms 
#in the house hence predictions made using this data will be wrong.
```

```python
x=data.columns
cor_mat = data[x].corr()
sns.heatmap(cor_mat)
```
