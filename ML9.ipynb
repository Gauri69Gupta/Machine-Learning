```python
import pandas as pd
import numpy as np
#data sets
d1=pd.read_csv("G:/train.csv")
d2=pd.read_csv("G:/test.csv")
dff=d1.append(d2,sort=False)
print(dff.head())
#2) Remove NaN values from the column ‘Survived’.
print("null values are",dff["Survived"].isnull().sum())
#Dropping null values
dff["Survived"].fillna(dff["Survived"].mean(),inplace = True)
print("null values after removing are",dff["Survived"].isnull().sum())
```

```python
#cabin 
dff["Has_Cabin"]=np.where(dff["Cabin"].isnull() == True ,"no", "yes")
dff.head()
```

```python
import re
#title
dff["title"] = dff.Name.apply(lambda x: re.search('([A-Z][a-z]+)\.', x).group(1))
dff["title"] = dff["title"].replace({'Mlle':'Mr', 'Mme':'Mrs', 'Ms':'Miss','Don':'Master'})
dff["title"] = dff["title"].replace(['Dona', 'Rev', 'Dr','Major', 'Lady', 'Sir', 'Col', 'Capt', 'Countess', 'Jonkheer'],'Special')
print(dff.head())
```

```python
print(dff.isnull().sum())
print("After putting values:",dff.fillna(dff["Age"].mean(),inplace=True))
print("AFter putting values in cabin"),dff.fillna(dff["Cabin"].mode(),inplace=True)
```

```python
for column in ["Age","Fare"]:
    dff[column].fillna(dff[column].mean(), inplace = True)
for column in ["Cabin","Embarked"]:
    dff[column].fillna(dff[column].mode(), inplace = True)
dff.isnull().sum()
```

```python
dff['Categorical_Age'] = pd.qcut(dff.Age, q=4, labels=False )
dff['Categorical_Fare']= pd.qcut(dff.Fare, q=4, labels=False)
dff.head()
```

```python
dff = dff.drop(["Cabin", "Name", "PassengerId", "Ticket", "Age", "Fare", "SibSp", "Parch"], axis=1)
dff.head()
```

```python
dff_dummy = pd.get_dummies(dff, drop_first=True)
dff_dummy.head()
```

```python
from sklearn.model_selection import train_test_split
x = dff[["Pclass"]]
y = dff["Survived"]

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state = 0)


#10) Use logistic regression by passing your X_train and y_train.
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
regressor.fit(x_train,y_train)
print("y_intercept:",regressor.intercept_)
print("slope:",regressor.coef_)

y_pred = regressor.predict(x_test)
da=pd.DataFrame({'actual':y_test,'predicted':y_pred})
print(da)
```
